```python
from LorentzTransformation import *

import matplotlib.pyplot as plt

import ipywidgets as widgets

from ipywidgets import interactive
#from ipywidgets import display
```

Lorentz transformations:
\begin{eqnarray}
ct^{\prime} & = & \gamma ( ct - \beta x ) \\
x^{\prime}  & = & \gamma ( x - \beta ct ) \\
y^{\prime}  & = & y \\
z^{\prime}  & = & z 
\end{eqnarray}

```python
def frames(beta=0.,s2=0,showInvariant=False,plot=False):
    fig = plt.gcf()
    fig.set_size_inches(14,7)

    plt.subplot(1,2,1)

    l = LorentzTransformation(beta=beta)
    # draw  Minkowski diagram

    # how does S see S'
    l.drawAxes(range=[-2,2,-2,2],plotInverse=False)
    # plot lightcone
    l.drawLightCone()

    if showInvariant:
        l.drawInvariant(s2=s2)

    plt.title('How does $S$ see $S^{\prime}$ for $\\beta$ = %0.2f' % beta)

    # how does S' see S
    plt.subplot(1,2,2)
    l.drawAxes(plotInverse=True)
    l.drawLightCone()
    
    if showInvariant:
        l.drawInvariant(s2=s2)
        
    plt.title('How does $S^{\prime}$ see $S$ for $\\beta$ = %0.2f' % beta)

    if plot:
        fig.savefig('frames.pdf')
```

```python
w = interactive(frames,beta=(-0.99,+0.99,0.01),s2=(-10,10,0.1),plot=False)
display(w)
```

```python
l = LorentzTransformation(gamma=2)


def ladder_in_barn(ct=0,plot=False):
    fig = plt.gcf()
    fig.set_size_inches(14,7)
    plt.subplot(1,2,1)

    plt.title('Ladder voor Baas')

    beta = l.beta
    # draw  Minkowski diagram
    l.drawAxes(range=[-1,3,-3,2],plotInverse=False)
    # plot lightcone
    l.drawLightCone()

    #barn
    #bx = np.array([0,0,0.25,0.75,1,1])
    #by = np.array([0,0.5,0.75,0.75,0.5,0])
    #by = by+ct
    #plt.plot(bx,by,color='green',linewidth=3)
    barn = Barn(L0=1,LT=l)
    barn.draw(ct,inRest=True)
    
    ladder = Ladder(L0=2,LT=l)
    ladder.draw(ct,inRest=False)

    plt.subplot(1,2,2)

    plt.title('Ladder voor Knecht')

    # draw  Minkowski diagram
    l.drawAxes(range=[-1,3,-3,2],plotInverse=True)
    # plot lightcone
    l.drawLightCone()
    
    barn.draw(ct,inRest=False)
    ladder.draw(ct,inRest=True)


    if plot:
        fig.savefig('ladders.pdf')
```

```python
w = interactive(ladder_in_barn,ct=(-2.5,+2.5,0.1),plot=False)
display(w)
```
